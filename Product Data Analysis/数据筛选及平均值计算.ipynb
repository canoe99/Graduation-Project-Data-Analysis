```python
import pandas as pd

# 读取Excel文档
file_path = 'D:\新建 XLSX 工作表.xlsx'
df = pd.read_excel(file_path)
```

```python
# 打印数据前几行和列名，确认读取情况
print("Excel数据预览：")
print(df.head())
print("列名：", df.columns)
```

```python
# 用户输入筛选条件
class_id = input("请输入班课id：")
grade = input("请输入年级：")
composition_title = input("请输入作文题目：")
```

```python
# 数据类型检查 (转化为字符串类型，确保比较正确)
df['班课id'] = df['班课id'].astype(str)
df['年级'] = df['年级'].astype(str)
df['作文题目'] = df['作文题目'].astype(str)
```

```python
# 筛选条件
filtered_df = df[(df['班课id'] == class_id) & (df['年级'] == grade) & (df['作文题目'] == composition_title)]
```

```python
# 需要计算平均值的列
columns_to_average = ['gpt打分', '开头score', '结尾score', '审题score',
                      '立意score', '详略score', '条理score', '选材score',
                      '想象score', '思想score', '结构score', '日记格式score',
                      '书信格式score', '倡议书格式score']
```

```python
filtered_df.fillna(0, inplace=True)
```

```python
# 计算每一列的平均值，并保留小数点后2位
average_scores = filtered_df[columns_to_average].mean().round(2)
```

```python
print("筛选后的平均分如下：")
print(average_scores)
```
