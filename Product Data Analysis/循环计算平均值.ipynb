```python
import pandas as pd

# 读取Excel文档
file_path = 'D:\新建 XLSX 工作表.xlsx'
df = pd.read_excel(file_path)
```

```python
# 数据类型确保转换为字符串，便于处理
df['班课id'] = df['班课id'].astype(str)
df['年级'] = df['年级'].astype(str)
df['作文题目'] = df['作文题目'].astype(str)
```

```python
# 获取唯一的班课id、年级和作文题目列表
class_ids = df['班课id'].unique()
grades = df['年级'].unique()
composition_titles = df['作文题目'].unique()
```

```python
# 需要计算平均值的列
columns_to_average = ['gpt打分', '开头score', '结尾score', '审题score',
                      '立意score', '详略score', '条理score', '选材score',
                      '想象score', '思想score', '结构score', '日记格式score',
                      '书信格式score', '倡议书格式score']
```

```python
# 将这些列确保转换为数值类型
for column in columns_to_average:
    df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
# 创建一个结果字典来存储每个组合的平均值
results = []
```

```python
# 循环遍历所有组合
for class_id in class_ids:
    for grade in grades:
        for composition_title in composition_titles:
            # 筛选条件
            filtered_df = df[(df['班课id'] == class_id) & 
                             (df['年级'] == grade) & 
                             (df['作文题目'] == composition_title)]
            
            # 检查过滤结果是否为空
            if not filtered_df.empty:
                # 填补NaN值；此处选择用列平均值填补NaN
                for column in columns_to_average:
                    if filtered_df[column].isnull().any():
                        filtered_df[column].fillna(filtered_df[column].mean(), inplace=True)
                
                # 计算每一列的平均值，并保留小数点后2位
                average_scores = filtered_df[columns_to_average].mean().round(2)
                
                # 保存结果
                results.append({
                    '班课id': class_id,
                    '年级': grade,
                    '作文题目': composition_title,
                    **average_scores.to_dict()
                })
```

```python
# 转换结果为DataFrame并展示
results_df = pd.DataFrame(results)
print("各组合的平均分如下：")
print(results_df)
```

```python
# 保存结果为新的Excel文档
output_file_path = 'filtered_results.xlsx'  # 输出文件的名称
results_df.to_excel(output_file_path, index=False)
```
