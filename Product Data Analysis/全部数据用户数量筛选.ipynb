```python
# 结果列表
result_data = []

def calculate_entries(data, class_ids, topics, grade):
    for class_id in class_ids:
        for topic, gpt_score_threshold in topics.items():
            filtered_data = data[(data['班课ID'] == class_id) & 
                                 (data['篇目'] == topic) & 
                                 (data['GPT打分'] < gpt_score_threshold)]
            result_data.append({
                '年级': grade,
                '班课ID': class_id,
                '篇目': topic,
                '条目数量': len(filtered_data)
            })
```

```python
# 计算所有年级数据
calculate_entries(data, class_ids_rule1, topics_rule1, '3年级')
calculate_entries(data, class_ids_rule2, topics_rule2, '3年级')
calculate_entries(data, class_ids_rule3, topics_rule3, '4年级')
calculate_entries(data, class_ids_rule4, topics_rule4, '4年级')
calculate_entries(data, class_ids_rule5, topics_rule5, '5年级')
calculate_entries(data, class_ids_rule6, topics_rule6, '5年级')
calculate_entries(data, class_ids_rule7, topics_rule7, '6年级')
calculate_entries(data, class_ids_rule8, topics_rule8, '6年级')
```

```python
# 将结果保存为Excel
result_df = pd.DataFrame(result_data)
output_file_path = 'D:\\result_all_grades.xlsx'
result_df.to_excel(output_file_path, index=False)
```
